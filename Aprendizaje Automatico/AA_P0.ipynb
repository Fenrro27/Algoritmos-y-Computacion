```python
# Importar las bibliotecas necesarias
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Cargar el dataset Iris
iris = load_iris()
X = iris.data  # Variables de entrada (features)
y = iris.target  # Etiquetas (target)

# Definir las particiones de los splits
splits = [(0.01, 0.99), (0.4, 0.6), (0.9, 0.1)]  # Training-Test (10-90, 50-50, 90-10)

# Inicializamos el clasificador
clf = DecisionTreeClassifier(random_state=42)

# Iteramos sobre cada partición y calculamos la accuracy
for train_size, test_size in splits:
    # Dividir los datos en conjuntos de entrenamiento y prueba según el tamaño especificado
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, train_size=train_size, random_state=42)
    
    # Entrenamos el árbol de decisión con los datos de entrenamiento
    clf.fit(X_train, y_train)
    
    # Predecimos las etiquetas para los datos de prueba
    y_pred = clf.predict(X_test)
    
    # Calculamos la accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # Mostramos los resultados
    print(f"Training-Test Split: {int(train_size*100)}-{int(test_size*100)}")
    print(f"Accuracy: {accuracy:.4f}\n")

    
```

```python
# Importar las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score

# Cargar el dataset Iris
iris = load_iris()
X = iris.data  # Variables de entrada (features)
y = iris.target  # Etiquetas (target)

# Definir las particiones de los splits
splits = [ (0.3,0.7),(0.8, 0.1), (0.9, 0.1)]  

# Inicializamos el clasificador
clf = DecisionTreeClassifier(random_state=42)

# Iteramos sobre cada partición y calculamos la accuracy
for train_size, test_size in splits:
    # Dividir los datos en conjuntos de entrenamiento y prueba según el tamaño especificado
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, train_size=train_size, random_state=42)
    
    # Entrenamos el árbol de decisión con los datos de entrenamiento
    clf.fit(X_train, y_train)
    
    # Predecimos las etiquetas para los datos de prueba
    y_pred = clf.predict(X_test)
    
    # Calculamos la accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # Mostramos los resultados
    print(f"Training-Test Split: {int(train_size*100)}-{int(test_size*100)}")
    print(f"Accuracy: {accuracy:.4f}\n")
    
    # Mostrar el árbol de decisión entrenado
    plt.figure(figsize=(20,10))
    plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
    plt.title(f"Árbol de Decisión - Split {int(train_size*100)}-{int(test_size*100)}")
    plt.show()
```

```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Cargar el dataset Iris
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Convertir las etiquetas a nombres de especies
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Contar instancias para cada valor de la clase
class_counts = iris_df['species'].value_counts()
print("Conteo de instancias por especie:")
print(class_counts)
print("\n")

# Configuración de los gráficos
sns.set(style='whitegrid')

# 1. Histograma de cada atributo
iris_df.iloc[:, :-1].hist(bins=10, figsize=(12, 8), grid=False)
plt.suptitle('Histogramas de Atributos del Dataset Iris', fontsize=16)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()

```
